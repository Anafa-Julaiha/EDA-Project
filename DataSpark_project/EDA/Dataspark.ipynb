```python
import pandas as pd
```

```python
# Read csv

data1 = pd.read_csv("C:\\Users\\Dell\\Downloads\\Customers.csv", encoding='cp1252')
data2 = pd.read_csv("C:\\Users\\Dell\\Downloads\\Exchange_Rates.csv", encoding= 'cp1252')
data3 = pd.read_csv("C:\\Users\\Dell\\Downloads\\Products.csv",encoding= 'cp1252')
data4 = pd.read_csv("C:\\Users\\Dell\\Downloads\\Stores.csv",encoding= 'cp1252')
data5 = pd.read_csv("C:\\Users\\Dell\\Downloads\\Sales.csv",encoding= 'cp1252')
```

#convert Dataframe

```python
customer = pd.DataFrame(data1)
customer
```

```python
exchange_rate = pd.DataFrame(data2)
exchange_rate.rename(columns = {'Currency':'Currency Code'}, inplace = True)
exchange_rate
```

```python
product = pd.DataFrame(data3)
product
```

```python
stores = pd.DataFrame(data4)
stores
```

```python
sales = pd.DataFrame(data5)
sales
```

```python

pd.set_option('display.max_columns', 31)  # Display 31 columns

#merge the data

customer2 = pd.merge(customer, sales, on='CustomerKey')
#customer3 =pd.merge(customer2,exchange_rate, on='Currency Code')
customer4 =pd.merge(customer2,product, on='ProductKey')
df = pd.merge(customer4,stores, on='StoreKey')
df.head()
```

```python
df = df.fillna(0)
df.head()
```

```python
df['Unit Cost USD'] = df['Unit Cost USD'].str.replace('$', '')
df['Unit Price USD'] = df['Unit Price USD'].str.replace('$', '')
```

```python
from sqlalchemy import create_engine
import psycopg2
from psycopg2.extensions import ISOLATION_LEVEL_AUTOCOMMIT

# Define your database connection details
host = "localhost"
port = "5432"  
database = "dataspark"
username = "postgres"
password = "1234"

# Create an engine instance
engine = create_engine(f'postgresql://{username}:{password}@{host}:{port}/{database}')

# Connect to the database using psycopg2
connection = psycopg2.connect(
    dbname=database, 
    user=username, 
    password=password, 
    host=host, 
    port=port
)
connection.set_isolation_level(ISOLATION_LEVEL_AUTOCOMMIT)

# Create a cursor object
writer = connection.cursor()

# Create the global_Data table
writer.execute("""
    CREATE TABLE global(
        CustomerKey INT,
        Gender VARCHAR(10),
        Name VARCHAR(255),
        City VARCHAR(255),
        StateCode VARCHAR(10),
        State_x VARCHAR(255),
        ZipCode VARCHAR(20),
        Country_x VARCHAR(255),
        Continent VARCHAR(50),
        BirthdayDate DATE,
        OrderNumber VARCHAR(255),
        DeliveryDate DATE,
        StoreKey INT,
        ProductKey INT,
        Quantity INT,
        CurrencyCode VARCHAR(10),
        ProductName VARCHAR(255),
        Brand VARCHAR(255),
        Color VARCHAR(50),
        UnitCostUSD FLOAT,
        UnitPriceUSD FLOAT,
        SubcategoryKey INT,
        Subcategory VARCHAR(255),
        CategoryKey INT,
        Category VARCHAR(255),
        Country_y VARCHAR(255),
        State_y VARCHAR(255),
        SquareMeter FLOAT,
        OpenDate DATE
    );
""")

# Close the connection
connection.close()
```
